## Extracting Rwanda GDP info 

```python
rwanda_gdp = gdp_data[gdp_data["Country Name"] == "Rwanda"]
```

```python
rwanda_gdp_df = rwanda_gdp.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="GDP"
)
```

## Female Primary Completion Rate 

```python
primary_completion_data_path = '/Users/judithdike/Documents/Fall 24/Viz Final/API_SE.PRM.CMPT.FE.ZS_DS2_en_csv_v2_847/API_SE.PRM.CMPT.FE.ZS_DS2_en_csv_v2_847.csv'
metadata_country_completion_path = '/Users/judithdike/Documents/Fall 24/Viz Final/API_SE.PRM.CMPT.FE.ZS_DS2_en_csv_v2_847/Metadata_Country_API_SE.PRM.CMPT.FE.ZS_DS2_en_csv_v2_847.csv'
metadata_indicator_completion_path = '/Users/judithdike/Documents/Fall 24/Viz Final/API_SE.PRM.CMPT.FE.ZS_DS2_en_csv_v2_847/Metadata_Indicator_API_SE.PRM.CMPT.FE.ZS_DS2_en_csv_v2_847.csv'
```

```python
primary_completion_data = pd.read_csv(primary_completion_data_path, skiprows=4)  # Skipping metadata rows
metadata_country_completion = pd.read_csv(metadata_country_completion_path)
metadata_indicator_completion = pd.read_csv(metadata_indicator_completion_path)
```

```python
rwanda_completion_data = primary_completion_data[primary_completion_data["Country Name"] == "Rwanda"]
```

```python
rwanda_completion_df = rwanda_completion_data.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="Completion Rate (Female)"
)
```

```python
rwanda_completion_df = rwanda_completion_df[["Year", "Completion Rate (Female)"]]
rwanda_completion_df["Year"] = pd.to_numeric(rwanda_completion_df["Year"], errors="coerce")
rwanda_completion_df["Completion Rate (Female)"] = pd.to_numeric(rwanda_completion_df["Completion Rate (Female)"], errors="coerce")
```

```python
# Merge Rwanda GDP data and Primary Completion Rate (Female) data on 'Year'
merged_data = pd.merge(
    rwanda_gdp_df,
    rwanda_completion_df,
    on="Year",
    how="outer"  # Keep all years from both datasets
)

# Print the first few rows of the merged dataset to inspect
print(merged_data.head())

# Check the structure of the merged dataset
merged_data.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(merged_data["Completion Rate (Female)"], merged_data["GDP"], alpha=0.7, color='teal')
plt.xlabel("Primary Completion Rate (Female) (%)")
plt.ylabel("GDP (Current USD)")
plt.title("Relationship Between Primary Completion Rate (Female) and GDP")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot GDP and Completion Rate (Female) with dual axes
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(merged_data["Year"], merged_data["GDP"], label="GDP (Current USD)", marker='o', color='blue')
ax1.set_xlabel("Year")
ax1.set_ylabel("GDP (Current USD)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(merged_data["Year"], merged_data["Completion Rate (Female)"], 
         label="Primary Completion Rate (Female)", marker='x', linestyle='--', color='orange')
ax2.set_ylabel("Primary Completion Rate (Female) (%)", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title("Rwanda's GDP and Primary Completion Rate (Female) Over Time")
fig.tight_layout()

plt.show()
```

```python
merged_data['GDP Growth Rate'] = merged_data['GDP'].pct_change() * 100
merged_data['Completion Rate Growth'] = merged_data['Completion Rate (Female)'].pct_change() * 100
```

```python
# Policy Milestones TEST ONLY
plt.figure(figsize=(12, 6))
plt.plot(merged_data["Year"], merged_data["GDP"], label="GDP (Current USD)", marker='o')
plt.plot(merged_data["Year"], merged_data["Completion Rate (Female)"], label="Primary Completion Rate (Female)", marker='x', linestyle='--')
```

```python
# Dual Axes Plot with Policy Milestones
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot GDP on the primary y-axis
ax1.plot(merged_data["Year"], merged_data["GDP"], label="GDP (Current USD)", marker='o', color='blue')
ax1.set_xlabel("Year")
ax1.set_ylabel("GDP (Current USD)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# secondary y-axis for Completion Rate
ax2 = ax1.twinx()
ax2.plot(merged_data["Year"], merged_data["Completion Rate (Female)"], 
         label="Primary Completion Rate (Female)", marker='x', linestyle='--', color='orange')
ax2.set_ylabel("Primary Completion Rate (Female) (%)", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# policy milestones (adjusted y-coordinates for alignment)
ax1.text(1995, 2e9, "1995: Education Reform", fontsize=10, color="red")
ax1.text(2000, 4e9, "2000: Free Primary Education", fontsize=10, color="red")
ax1.text(2010, 8e9, "2010: Gender Parity Initiative", fontsize=10, color="red")

plt.title("Rwanda's GDP and Primary Completion Rate (Female) with Policy Milestones")
fig.tight_layout()

plt.show()
```
