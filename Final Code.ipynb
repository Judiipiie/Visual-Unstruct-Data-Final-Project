```python
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objs as go
import pandas as pd

# Initialize Dash app
app = Dash(__name__)

policies = {
    "1994: Post-Genocide Reconstruction": 1994,
    "2003: Constitutional Quota for Women": 2003,
    "2004: Girls’ Education Task Force": 2004,
    "2008: Girls’ Education Policy": 2008,
    "2010: National Education Policy": 2010,
    "2012: Commonwealth Award": 2012,
    "2014: Akilah Institute": 2014,
    "2016: University Gender Policy": 2016,
    "2017: National Strategy for Transformation": 2017,
    "2021: Revised Gender Policy": 2021,
    "2024: 30 Years of Women's Empowerment": 2024
}


merged_data = pd.DataFrame({
    "Year": list(range(1990, 2025)),
    "GDP": [1e9 + (i * 2e9) for i in range(35)],
    "Completion Rate (Female)": [40 + (i * 1.5) for i in range(35)]
})


app.layout = html.Div([
    html.H1("Interactive Visualization: Rwanda's GDP and Education Milestones"),
    
    dcc.Dropdown(
        id="policy-selector",
        options=[{"label": k, "value": v} for k, v in policies.items()],
        value=1994,
        placeholder="Select a Policy Milestone"
    ),
    
    dcc.Graph(id="dual-axis-plot")
])


@app.callback(
    Output("dual-axis-plot", "figure"),
    Input("policy-selector", "value")
)
def update_graph(selected_year):
    # Base plot
    fig = go.Figure()


    fig.add_trace(go.Scatter(
        x=merged_data["Year"],
        y=merged_data["GDP"],
        mode="lines+markers",
        name="GDP (Current USD)",
        line=dict(color="blue"),
        marker=dict(size=8)
    ))


    fig.add_trace(go.Scatter(
        x=merged_data["Year"],
        y=merged_data["Completion Rate (Female)"],
        mode="lines+markers",
        name="Primary Completion Rate (Female)",
        line=dict(color="darkorange", dash="dash"),
        marker=dict(size=8)
    ))


    if selected_year in merged_data["Year"].values:
        gdp_value = merged_data.loc[merged_data["Year"] == selected_year, "GDP"].values[0]
        completion_value = merged_data.loc[merged_data["Year"] == selected_year, "Completion Rate (Female)"].values[0]

        fig.add_trace(go.Scatter(
            x=[selected_year],
            y=[gdp_value],
            mode="markers+text",
            name="Selected Policy Year (GDP)",
            text=[f"GDP: {gdp_value:,.0f}"],
            textposition="top center",
            marker=dict(color="red", size=12, symbol="star")
        ))

        fig.add_trace(go.Scatter(
            x=[selected_year],
            y=[completion_value],
            mode="markers+text",
            name="Selected Policy Year (Completion Rate)",
            text=[f"Rate: {completion_value:.2f}%"],
            textposition="top center",
            marker=dict(color="green", size=12, symbol="diamond")
        ))

    fig.update_layout(
        title="Rwanda's GDP and Primary Completion Rate (Female) with Policy Milestones",
        xaxis=dict(title="Year"),
        yaxis=dict(title="GDP (Current USD)", side="left"),
        yaxis2=dict(title="Completion Rate (Female) (%)", overlaying="y", side="right"),
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
        template="plotly_white"
    )
    return fig


if __name__ == "__main__":
    app.run_server(debug=True)
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objs as go
import pandas as pd

# Initialize Dash app
app = Dash(__name__)

# Sample data for policies
policies = {
    "1994: Post-Genocide Reconstruction": {
        "year": 1994,
        "description": "After the genocide, women played a pivotal role in rebuilding Rwanda's socio-economic structures."
    },
    "2003: Constitutional Quota for Women": {
        "year": 2003,
        "description": "Introduced 30% representation quota for women in all decision-making bodies."
    },
    "2004: Girls’ Education Task Force": {
        "year": 2004,
        "description": "Promoted gender equality in education through targeted initiatives."
    },
    "2008: Girls’ Education Policy": {
        "year": 2008,
        "description": "Addressed barriers hindering girls' education, ensuring equal opportunities."
    },
    "2010: National Education Policy": {
        "year": 2010,
        "description": "Focused on inclusive education to drive national development."
    },
    "2012: Commonwealth Award": {
        "year": 2012,
        "description": "Rwanda was honored for increasing equal access to basic education for boys and girls."
    },
    "2014: Akilah Institute": {
        "year": 2014,
        "description": "Established Rwanda's first all-female college to support women's leadership."
    },
    "2016: University Gender Policy": {
        "year": 2016,
        "description": "Allocated 50% of university positions to women, promoting gender parity."
    },
    "2017: National Strategy for Transformation": {
        "year": 2017,
        "description": "Focused on equitable and quality education for lifelong learning."
    },
    "2021: Revised Gender Policy": {
        "year": 2021,
        "description": "Updated to ensure gender equality principles across all sectors."
    },
    "2024: 30 Years of Women's Empowerment": {
        "year": 2024,
        "description": "Celebration of women's contribution to Rwanda's transformation."
    }
}

merged_data = pd.DataFrame({
    "Year": list(range(1990, 2025)),
    "GDP": [1e9 + (i * 2e9) for i in range(35)],
    "Completion Rate (Female)": [40 + (i * 1.5) for i in range(35)]
})

# App layout
app.layout = html.Div([
    html.H1("Interactive Dash App: Rwanda's GDP and Education Milestones"),
    html.P("Select a policy milestone to view details on the graph:"),
    
    dcc.Dropdown(
        id="policy-selector",
        options=[{"label": k, "value": v["year"]} for k, v in policies.items()],
        value=1994,
        placeholder="Select a Policy Milestone"
    ),
    
    dcc.Graph(id="interactive-plot"),
    
    html.Div(id="policy-details", style={"font-size": "16px", "margin-top": "20px"})
])

# Callback 
@app.callback(
    [Output("interactive-plot", "figure"), Output("policy-details", "children")],
    [Input("policy-selector", "value")]
)
def update_graph(selected_year):
    
    fig = go.Figure()

    
    fig.add_trace(go.Scatter(
        x=merged_data["Year"],
        y=merged_data["GDP"],
        mode="lines+markers",
        name="GDP (Current USD)",
        line=dict(color="blue"),
        marker=dict(size=8)
    ))


    fig.add_trace(go.Scatter(
        x=merged_data["Year"],
        y=merged_data["Completion Rate (Female)"],
        mode="lines+markers",
        name="Primary Completion Rate (Female)",
        line=dict(color="darkorange", dash="dash"),
        marker=dict(size=8)
    ))


    policy_detail = ""
    if selected_year:
        policy_info = next((info for key, info in policies.items() if info["year"] == selected_year), None)
        if policy_info:
            policy_detail = f"**{policy_info['year']}**: {policy_info['description']}"
            
            gdp_value = merged_data.loc[merged_data["Year"] == selected_year, "GDP"].values[0]
            completion_value = merged_data.loc[merged_data["Year"] == selected_year, "Completion Rate (Female)"].values[0]

            fig.add_trace(go.Scatter(
                x=[selected_year],
                y=[gdp_value],
                mode="markers+text",
                name="Selected Policy Year (GDP)",
                text=[f"GDP: {gdp_value:,.0f}"],
                textposition="top center",
                marker=dict(color="red", size=12, symbol="star")
            ))

            fig.add_trace(go.Scatter(
                x=[selected_year],
                y=[completion_value],
                mode="markers+text",
                name="Selected Policy Year (Completion Rate)",
                text=[f"Rate: {completion_value:.2f}%"],
                textposition="top center",
                marker=dict(color="green", size=12, symbol="diamond")
            ))


    fig.update_layout(
        title="Rwanda's GDP and Primary Completion Rate (Female) with Policy Milestones",
        xaxis=dict(title="Year"),
        yaxis=dict(title="GDP (Current USD)", side="left"),
        yaxis2=dict(title="Completion Rate (Female) (%)", overlaying="y", side="right"),
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
        template="plotly_white"
    )
    return fig, policy_detail

# Run the app
if __name__ == "__main__":
    app.run_server(debug=True)
```
