## Extracting Rwanda GDP info 

```python
rwanda_gdp = gdp_data[gdp_data["Country Name"] == "Rwanda"]
```

```python
rwanda_gdp_df = rwanda_gdp.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="GDP"
)
```

```python
rwanda_gdp_df = rwanda_gdp_df[["Year", "GDP"]]
rwanda_gdp_df["Year"] = pd.to_numeric(rwanda_gdp_df["Year"], errors="coerce")
rwanda_gdp_df["GDP"] = pd.to_numeric(rwanda_gdp_df["GDP"], errors="coerce")
```
